```python
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, variant="fp16")
pipe.to("cuda")
```

```python
prompt = "Astronaut dressed like Napoleon standing in a desert mountainous landscape, cold color palette, muted colors, detailed, 8k"
image = pipe(prompt=prompt).images[0]
image
```
